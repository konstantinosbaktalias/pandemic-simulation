```python
import numpy as np
import pandas as pd
import random, time
import matplotlib.pyplot as plt

from pandemicsimulition import Pandemic
```

```python
pandemic = Pandemic(
    area_shape=(1000, 1000), 
    num_of_people=1000,
    infection_rate=0.3,
    fatality_rate=0.2,
    recover_period=5,
    death_period=5,
    intelligent=False
    )
```

```python
pandemic.run(animate=True)
```

```python
pandemic.plot_resaults()
```
